```python
import os
import torch
import pickle
import transformers
import pandas as pd

from tqdm import tqdm
from google.colab import drive
from transformers import AutoTokenizer
```

```python
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab Notebooks/wallpaper-defects-qa/
```

```python
test_file_name = 'test.csv'
test_df = pd.read_csv('data/open/' + test_file_name)
test_df.head()
```

```python
q_list = test_df['질문'].values.tolist()
```

```python
q_list[0]
```

# beomi/llama-2-ko-7b

```python
model_name = "beomi/llama-2-ko-7b"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    do_sample=True,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=512,
)
```

```python
# random index
random_idx = [58, 41, 126, 9, 113, 107, 31, 65, 62, 16]
```

```python
answers = []

q_list_random = [q_list[i] for i in random_idx]

for q in tqdm(q_list_random):
    answer = pipeline(q)
    answers.append(answer)
    print(answer)
```

```python
answers
```

```python
with open(f'outputs/output_for_test_{os.path.basename(model_name)}.pickle', 'wb') as f:
    pickle.dump(answers, f)
```
